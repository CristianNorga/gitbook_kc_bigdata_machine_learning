### Divide y vencerás
- Divide un problema complejo en problemas más pequeños y manejables.
- Resuelve cada problema pequeño por separado.
- Combina las soluciones de los problemas pequeños para resolver el problema original.

### Transforma y vencerás
- Transforma un problema difícil en uno más fácil de resolver.
- Utiliza analogías o ejemplos más simples para entender el problema.
- Aplica técnicas de simplificación para reducir la complejidad del problema.

### KISS: Keep It Simple, Stupid!
- Mantén tus soluciones simples y directas.
- Evita la sobrecomplicación y los enfoques innecesariamente complejos.
- Busca la elegancia en la simplicidad.

> YAGNI: You Aren't Gonna Need It! (no lo vas a necesitar).

### DRY: Don't Repeat Yourself!
- Evita la duplicación de código y lógica.
- Extrae funciones o métodos reutilizables para evitar la repetición.
- Mantén la lógica de negocio en un solo lugar.

### SRP: Single Responsibility Principle
- Cada módulo o clase debe tener una única responsabilidad.
- Si una clase o módulo tiene más de una responsabilidad, es más difícil de entender y mantener.
- La separación de responsabilidades facilita la reutilización y la prueba de componentes individuales.

### Cerveza para todos o para nadie
- Evita los casos especiales en tu código.
- Si encuentras un caso especial, trata de generalizarlo o eliminarlo.
- Mantén la coherencia en tu código para facilitar su comprensión y mantenimiento.

### Vale más una estructura de datos que diez algoritmos
- Elige estructuras de datos adecuadas para resolver problemas.
- Utiliza las estructuras de datos para simplificar la lógica de tu código.
- A veces, una buena estructura de datos puede reemplazar la necesidad de algoritmos complejos.

### Leña al mono hasta que pase los tests
- Escribe pruebas automatizadas para tu código.
- Asegúrate de que todas las pruebas pasen antes de realizar cambios en el código.
- Si una prueba falla, corrige el código hasta que la prueba pase.
