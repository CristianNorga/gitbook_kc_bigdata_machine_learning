## Abstracción de ambitos

- Definición de visibilidad y ámbito
  - Visibilidad: alcance de una variable o función
  - Ámbito: contexto en el que una variable o función es accesible

- Ámbito local
  - Variables definidas dentro de una función
  - No accesibles fuera de la función
- Ámbito global
  - Variables definidas fuera de cualquier función
  - Accesibles desde cualquier parte del código
- Ámbito de clase
  - Variables y métodos definidos dentro de una clase
  - Accesibles a través de instancias de la clase
- Ámbito de módulo
  - Variables y funciones definidas en un archivo de módulo
  - Accesibles mediante importación del módulo
